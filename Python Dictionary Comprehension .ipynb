Question 5: Write a Python program to create a dictionary with keys as numbers from 1 to 10 and values as their even/odd status using dictionary comprehension

- `sample input : `
    - `var1 = 'listen'`
    - `var2 = 'silent'`
- `sample output :`
    - `result : both are anagram because characters in var2 is in var1`

```python
# Please write The solution Here for above question 5:
even_odd_dict={num:"even" if num%2==0 else "odd" for num in range(1,11)}
print(even_odd_dict)
var1='listen'
var2='silent'
if sorted(var1)==sorted(var2):
    result="both are anagram because characters in var2 is in var1"
else:
    result="not a anagram"
print("result:",result)
```

Question 6: Write a Python program to create a dictionary with keys as words from a given sentence and values as their lengths using dictionary comprehension.



- `sample input : `
    - `var = 'python'`
- `sample output :`
    - `result : pythn , because vowels are [AEIOUaeiou]`

```python
# Please write The solution Here for above question 6:
input_sentence='python is amazing'
word_length_dict={word:len(word) for word in input_sentence.split()}
print(word_length_dict)
input_word='python'
output_word=''.join(char for char in input_word if char.lower() not in 'aeiou')
print("result:",output_word)
```

Question 7: Write a Python program to create a dictionary with keys as letters from the alphabet and values as their ASCII values using dictionary comprehension..

- `sample input : `
    - `var = 'python is Easy Language'`
- `sample output :`
    - `result : Language`

```python
# Please write The solution Here for above question 7:
input_str='python is Easy Language'
ascii_dict={char:ord(char) for char in input_str if char.isalpha()}
print(ascii_dict)
words=input_str.split()
language_word=[word for word in words if word.lower()=='language']
print("result:",''.join(language_word))
```

Question 8: Write a Python program to create a dictionary with keys as numbers from 1 to 10 and values as their factorial using dictionary comprehension.


- `sample input : `
    - `var = 'python is easy'`
- `sample output :`
    - `result : Python Is Easy`

```python
# Please write The solution Here for above question 8:
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)
factorial_dict={num:factorial(num) for num in range(1,11)}
print(factorial_dict)
input_str='python is easy'
capitalized_str=input_str.title()
print("result:",capitalized_str)
```
