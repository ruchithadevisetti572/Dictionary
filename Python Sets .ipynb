### Question:4 Write a Python program to check if one set is a subset of another.

- `Sample Input: {1, 2, 3}, {1, 2, 3, 4, 5}`
- `Sample Output: True`

```python
# Please write The solution Here for above question 5:

set1={1,2,3}
set2={1,2,3,4,5}
result=set1.issubset(set2)
print("result:",result)
```
